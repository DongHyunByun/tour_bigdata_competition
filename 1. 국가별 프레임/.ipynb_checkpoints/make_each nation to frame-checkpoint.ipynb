```python
import csv
while(1):
    fileMatrix=[]
    fileName=input()
    if fileName=="exit":
        break
    with open(fileName+".csv",'r') as fileOpen:
        for i in fileOpen :
            tempL=i.strip("\n").split(",")
            if (len(tempL)!=11) or ("NA" in tempL[10]) or ("서울" not in tempL[2]):
                continue
            fileMatrix.append(tempL)
    #att =>지역[0] 소분류[1~12] , 여가[13~14], 평일주말[15~16], 활휴취[17~19]
    att=["지역","교통_총금액","교통_총횟수","쇼핑_총금액","쇼핑_총횟수","의료_총금액","의료_총횟수",
        "숙박_총금액","숙박_총횟수","체험_총금액","체험_총횟수","스포츠_총금액","스포츠_총횟수",
        "여가_총금액","여가_총횟수","평일","주말","활동","휴식","취침"]
    #시군구 분류
    local=set()
    for i in fileMatrix[1:]:
        local.add(i[3])
    #완성될 frame
    frame=[att]
    for i in local:
        frame.append([i]+[0 for i in range(19)])
    
    #파일 contents 읽기
    for i in fileMatrix[1:]:
        if "NA" in i[10]:
            continue
        indx=0
        for j in range(len(frame)):
            if frame[j][0]==i[3]:
                indx=j
                break
        #목적대분류(금액과 총횟수)
        if i[4] in frame[0][1]:
            frame[indx][1]+=int(i[9])
            frame[indx][2]+=int(i[10])
        elif i[4] in frame[0][3]:
            frame[indx][3]+=int(i[9])
            frame[indx][4]+=int(i[10])
        elif i[4] in frame[0][5]:
            frame[indx][5]+=int(i[9])
            frame[indx][6]+=int(i[10])
        elif i[4] in frame[0][7]:
            frame[indx][7]+=int(i[9])
            frame[indx][8]+=int(i[10])
        elif i[4] in frame[0][9]:
            frame[indx][9]+=int(i[9])
            frame[indx][10]+=int(i[10])
        elif i[4] in frame[0][11]:
            frame[indx][11]+=int(i[9])
            frame[indx][12]+=int(i[10])
        #오전 오후
        if i[7]=="WHITE":
            frame[indx][15]+=1
        else:
            frame[indx][16]+=1
        #활휴취
        if i[8]=="활동":
            frame[indx][17]+=1
        elif i[8]=="휴식":
            frame[indx][18]+=1
        else:
            frame[indx][19]+=1
    #여가 합
    for i in frame[1:]:
        i[13]=i[3]+i[9]+i[11]
        i[14]=i[4]+i[10]+i[12]
        
    print(frame)
    with open(fileName+"W.csv","w",newline='') as fileWrite:
        csvWriter=csv.writer(fileWrite)
        for i in frame:
            csvWriter.writerow(i)
```
