```python
import matplotlib.pyplot as plt

fileMatrixV=[]
with open("공통att방문횟수.csv","r",encoding="utf-8") as fileOpen:
    for i in fileOpen:
        fileMatrixV.append(i.strip("\n").split(","))
        
fileMatrixS=[]
with open("공통att만족도.csv","r",encoding="utf-8") as fileOpen:
    for i in fileOpen:
        fileMatrixS.append(i.strip("\n").split(","))


fig,ax=plt.subplots()
for i in range(1,8):
    ax.plot(list(map(float,fileMatrixS[i][1:])),list(map(float,fileMatrixV[1][1:])),marker="o",linestyle="",label=fileMatrixV[i][0])
ax.legend(fontsize=7,loc="upper left")
plt.xlabel("satisfy")
plt.ylabel("visit")
plt.show()

print("[국가별 local A와 local B]")
print("*local A : 만족도와 방문율이 모두 높은 지역 (만족도*방문율)")
print("*local B : 상대적으로 만족도는 높으나 방문율은 낮은 지역 (만족도/방문율)")
print()
print()
print("-----------------------------------------------------------------")

for i in range(1,8):
    print("                           <국가 : ",fileMatrixS[i][0],">")
    groupA=[]
    groupB=[]
    for j in range(1,len(fileMatrixS[0])):            
        if fileMatrixV[i][j]=="0" or fileMatrixS[i][j]=="0" :
            continue    
        else:
            groupA.append([float(fileMatrixS[i][j])*float(fileMatrixV[i][j]),fileMatrixS[0][j]])
            groupB.append([float(fileMatrixS[i][j])/float(fileMatrixV[i][j]),fileMatrixS[0][j]])
    
    groupA.sort(reverse=True)
    groupB.sort(reverse=True)
    print(len(groupA))
    print("  [LOCAL A]")
    count=0
    for A in groupA[:5]:
        print(count+1,"번째 : ",A[1])
        count+=1
    print("  [LOCAL B]")
    count=0
    for B in groupB:
        if count==5:
            break
        isSame=False
        for A in groupA[:5]:
            if B[1]==A[1]:
                isSame=True
                break
        if not isSame:
            print(count+1,"번째 : ",B[1])
            count+=1
    print("-----------------------------------------------------------------")
```
